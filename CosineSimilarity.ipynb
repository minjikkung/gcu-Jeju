```python
import pandas as pd
import numpy as np
from numpy import dot

from numpy.linalg import norm
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_excel("jeju place5.xlsx",na_filter=True)
del df['Unnamed: 0']
```

```python
df
```

```python
dfe=df.groupby(df['콘텐츠분류']).get_group('음식점')
dfs=df.groupby(df['콘텐츠분류']).get_group('숙박')
dfg=df.groupby(df['콘텐츠분류']).get_group('관광지')
```

```python
dfe=dfe.dropna(subset=['리뷰'])
dfs=dfs.dropna(subset=['리뷰'])
dfg=dfg.dropna(subset=['리뷰'])
```

```python
from konlpy.tag import Kkma
from collections import Counter
kkma=Kkma()
```

```python
list=[]
for i in range(0,len(dfs)):
    list+=kkma.nouns(dfs.iloc[i,-4])
count=Counter(list)
count.most_common(50)
```

```python
list=[]
for i in range(0,len(dfg)):
    list+=kkma.nouns(dfg.iloc[i,-4])
count=Counter(list)
count.most_common(50)
```

```python
list=[]
for i in range(0,len(dfe)):
    list+=kkma.nouns(dfe.iloc[i,-1])
count=Counter(list)
count.most_common(50)
```

```python
count.most_common(10)
```

```python
# 코드 수정
# 유사도 점수에 따라 추천
# 해시태그 빈도 
tfidf_vectorizer = TfidfVectorizer()

sentence=['맛집 추천 최고 친절 강추']

cos_sim=[]

for i in range(0,len(dfe)):
    sentence.append(dfe.iloc[i,-4])
    
tfidf_matrix = tfidf_vectorizer.fit_transform(sentence) 

for i in range(0,len(dfe)):
    a=i+1;b=i+2
    cos_similar=cosine_similarity(tfidf_matrix[0:1],tfidf_matrix[a:b])
    print("기준 문장과 ",i,"번째 문장과의 코사인 유사도: ",cos_similar)
    cos_sim.append(cos_similar)
```

```python
c=np.array(cos_sim,dtype=float)
c=c.tolist()
arr=[]
for i in range(0,len(c)):
    arr.append(c[i][0][0])
dfe['코사인']=arr
```

```python
cos_sim
```

```python
dfe.sort_values('코사인',ascending=False)
```
