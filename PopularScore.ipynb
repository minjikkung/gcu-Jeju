```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_excel("jeju place5.xlsx")
del df['Unnamed: 0']
```

```python
df=df.iloc[:,0:-2]
```

```python
def normalize(arr):
    normal = []
    
    for i in arr:
        num = (i - min(arr)) / (max(arr) - min(arr))
        normal.append(num)
    
    return normal
```

```python
arr=df['인기점수'].to_list()
```

```python
arr=normalize(arr)
```

```python
df['정규화']=0.0
for i in range(0,len(df)):
    df.iloc[i,-1]=arr[i]
```

```python
df['가중치']=0.0
df['가중치']=df['정규화']*df['코사인']
```

```python
df.sort_values(['가중치'],ascending=False)
```

```python
df.to_excel('jeju place5.xlsx')
```
