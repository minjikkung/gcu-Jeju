```python
import pandas as pd
import collections
```

```python
df=pd.read_excel("visit jeju total.xlsx")
del df['Unnamed: 0']
```

```python
df.iloc[0,13]
```

```python
list=[]
for i in range(0,len(df)):
    list+=df.iloc[i,13].split(' ')
```

```python
list
```

```python
counts=collections.Counter(list)
```

```python
counts.most_common(15)
```

```python
dfg=df.groupby('콘텐츠분류').get_group('관광지') # g 관광지
dfe=df.groupby('콘텐츠분류').get_group('음식점') # e 음식점
dfs=df.groupby('콘텐츠분류').get_group('숙박') # s 숙박
```

```python
listg=[]
for i in range(0,len(dfg)):
    listg+=dfg.iloc[i,13].split(' ')
```

```python
liste=[]
for i in range(0,len(dfe)):
    liste+=dfe.iloc[i,13].split(' ')
```

```python
lists=[]
for i in range(0,len(dfs)):
    lists+=dfs.iloc[i,13].split(' ')
```

```python
countsg=collections.Counter(listg)
countse=collections.Counter(liste)
countss=collections.Counter(lists)
```

```python
countsg.most_common(15)
```
